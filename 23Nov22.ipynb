```python
def OverView(CE_Sell,PE_Sell,CE_Buy,PE_Buy):
    try:
        Max_Profit=round((CE_Sell+PE_Sell-CE_Buy-PE_Buy)*25,2)
        Max_Loss=round((CE_Sell+PE_Sell-CE_Buy-PE_Buy-200)*25,2)
        List=[Max_Profit,Max_Loss,round(-Max_Loss/Max_Profit,2)]
    except Exception as e:
        return [str(e),str(e),str(e)]
    else:
        return List
```

```python
def CallPrice(StrikePrice,Data):
    try:
        df=Data[Data["CE.strikePrice"]==StrikePrice]["CE.lastPrice"].iloc[0]
    except Exception as e:
        return np.NaN
    else:
        return df

def PutPrice(StrikePrice,Data):
    try:
        df=Data[Data["PE.strikePrice"]==StrikePrice]["PE.lastPrice"].iloc[0]
    except Exception as e:
        return np.NaN
    else:
        return df

def CE_OpenInterest(StrikePrice,Data):
    try:
        df=Data[Data["CE.strikePrice"]==StrikePrice]["CE.openInterest"].iloc[0]
    except Exception as e:
        return np.NaN
    else:
        return df
    
def PE_OpenInterest(StrikePrice,Data):
    try:
        df=Data[Data["PE.strikePrice"]==StrikePrice]["PE.openInterest"].iloc[0]
    except Exception as e:
        return np.NaN
    else:
        return df

def PE_underlyingValue(Data):
    try:
        df=Data["PE.underlyingValue"].iloc[0]
    except Exception as e:
        return np.NaN
    else:
        return df
```

```python
my_columns = ["CE StrikePrice", "CE OpenInterest",'PE StrikePrice',"PE OpenInterest","CE StrikePrice 1","CE OpenInterest 1", 'PE StrikePrice 1',"PE OpenInterest 1","PE UnderlyingValue",'Max Profit', 'Max Loss', "Risk/Reward"]
final_dataframe = pd.DataFrame(columns = my_columns)


for diff in range(100,400,100):
#     print(diff)
    for i in df_call_1["CE.strikePrice"]:
        l=[]
        l.append(i)
        l.append(CE_OpenInterest(i,df_data_n))
        for j in df_put_1["PE.strikePrice"]:
            l.append(j)
            l.append(PE_OpenInterest(j,df_data_n))

            l.append(l[0]+diff)
            l.append(CE_OpenInterest(l[0]+diff,df_data_n))

            l.append(l[2]-diff)
            l.append(PE_OpenInterest(l[2]-diff,df_data_n))

            l.append(PE_underlyingValue(df_data_n))

            CE_Sell=CallPrice(l[0],df_data_n)
            PE_Sell=PutPrice(l[2],df_data_n)
            CE_Buy=CallPrice(l[4],df_data_n)
            PE_Buy=PutPrice(l[6],df_data_n)
            l=l+OverView(CE_Sell,PE_Sell,CE_Buy,PE_Buy)

            final_dataframe = final_dataframe.append(
                                            pd.Series(l,index = my_columns), 
                                            ignore_index = True)
            l=[i,CE_OpenInterest(i,df_data_n)]
```

```python
final_dataframe['Risk/Reward'] = pd.to_numeric(final_dataframe['Risk/Reward'], errors="coerce")
final_dataframe.dropna(how="any", inplace=True)
```
